####**1.2.1.3 Fatiamento e indexação de tensores**


#####**1.2.1.3.1 Fatiamento**

Os métodos de fatiamento de tensores incluem:
*   **[start:end]**: extrai uma fatia de dados da posição inicial até a posição final do tensor.
*   **[start:end:step]** ou **[::step]**: extrai uma fatia de dados em um intervalo de **step** desde o início posição até a posição final do tensor.
*   **[::-1]**: divide os dados do último elemento.
*   **'...'**: indica uma fatia de dados de qualquer comprimento.

```python
#Crie um tensor quadridimensional. O tensor contém quatro imagens. O tamanho de cada imagem é 100 x 100 x 3.
tensor_h = tf.random.normal([4,100,100,3])
tensor_h
```

```python
#Extraia a primeira imagem.
tensor_h[0,:,:,:]
```

```python
#Extraia uma fatia em um intervalo de duas imagens.
tensor_h[::2,...]
```

```python
#Fatie dados do último elemento.
tensor_h[::-1]
```

#####**1.2.1.3.2 Indexação**

O formato básico de um índice é o seguinte: **a[d1][d2][d3]**.

```python
#Obtenha o pixel na posição [20,40] no segundo canal da primeira imagem.
tensor_h[0][19][39][1]
```

Se os índices de dados a serem extraídos não forem consecutivos, **tf.gather** e **tf.gather_nd** serão comumente usados para extração de dados no TensorFlow.

Para extrair dados de uma dimensão específica:
t**f.gather(params, indices,axis=None)**:
*   **params**: tensor de entrada.
*   **indices**: índice dos dados a serem extraídos.
*   **axis**: dimensão dos dados a serem extraídos.

```python
#Extraia a primeira, segunda e quarta imagens do tensor_h ([4.100.100,3]).
indices = [0,1,3]
tf.gather(tensor_h,axis=0,indices=indices,batch_dims=1)
```

**tf.gather_nd** permite que os índices sejam executados em múltiplas dimensões:
**tf.gather_nd(params,indices)**:
*   **params**: tensor de entrada.
*   **indices**: índice dos dados a serem extraídos, que geralmente é uma lista multidimensional.

```python
#Extraia o pixel em [1,1] da primeira dimensão da primeira imagem e o pixel em [2,2] da
#primeira dimensão da segunda imagem em tensot_h ([4.100.100,3]).
indices = [[0,1,1,0],[1,2,2,0]]
tf.gather_nd(tensor_h,indices=indices)
```

#####**1.2.1.4.3 Expansão**

**tf.expand_dims(input,axis,name=None)**:
*   **input**: tensor de entrada
*   **axis**: adiciona uma dimensão após a dimensão do eixo. Quando o número de dimensões dos dados de entrada são D, o eixo deve estar no intervalo de **[–(D + 1), D]** (incluído). Um valor negativo indica a adição de uma dimensão na ordem inversa.

```python
#Gere um tensor 100 x 100 x 3 para representar uma imagem colorida de três canais 100 x 100.
expand_sample_1 = tf.random.normal([100,100,3], seed=1)
print("tamanho dos dados originais:",expand_sample_1.shape)
print("adicionar uma dimensão antes da primeira dimensão (eixo=0): ",tf.expand_dims(expand_sample_1,axis=0).shape)
print("adicionar uma dimensão antes da segunda dimensão (eixo=1): ",tf.expand_dims(expand_sample_1,axis=1).shape)
print("adicionar uma dimensão após a última dimensão (axis=-1): ",tf.expand_dims(expand_sample_1, axis=-1).shape)
```

#####**1.2.1.4.5 Transposição**

**tf.transpose(a,perm=None,conjugate=False,name='transpose')**:
*   **a**: tensor de entrada
*   **perm**: sequência de tamanho de tensor, geralmente usada para transpor matrizes de alta dimensão
*   **conjugate**: indica transposição de número complexo.
*   **name**: nome do tensor

```python
#O problema de transposição da baixa dimensão é simples. A entrada precisa ser chamada transpondo tensores.
tf.transpose
trans_sample_1 = tf.constant([1,2,3,4,5,6],shape=[2,3])
print("tamanho dos dados originais:",trans_sample_1.shape)
transposed_sample_1 = tf.transpose(trans_sample_1)
print("tamanho dos dados transpostos:",transposed_sample_1.shape)
```

**perm** é necessário para transposição de dados de alta dimensão e indica a sequência de dimensão do tensor de entrada. A sequência de dimensão original de um tensor tridimensional é **[0, 1, 2]** (**perm)**, indicando o comprimento, largura e altura de dados de alta dimensão, respectivamente. As dimensões de dados podem ser transpostas alterando a sequência de valores em **perm**.

```python
#Gere um tensor $ x 100 x 200 x 3 para representar quatro imagens coloridas de três canais de 100 x 200.
trans_sample_2 = tf.random.normal([4,100,200,3])
print("tamanho dos dados originais:",trans_sample_2.shape)
#Troque o comprimento e a largura das quatro imagens. O valor de permissão original é [0,1,2,3] e o o novo valor de permissão é [0,2,1,3].
transposed_sample_2 = tf.transpose(trans_sample_2,[0,2,1,3])
print("tamanho dos dados transpostos:",transposed_sample_2.shape)
```

####**1.2.1.6 Operações Aritméticas Baseadas em Dimensões**

No TensorFlow, uma série de operações de  **tf.reduce_\*** reduz as dimensões do tensor. As séries de operações podem ser realizadas em elementos dimensionais de um tensor, por exemplo, calculando o valor médio por linha e calculando um produto de todos os elementos no tensor. As operações comuns incluem **tf.reduce_sum** (adição), **tf.reduce_prod** (multiplicação), **tf.reduce_min** (mínimo), **tf.reduce_max** (máximo), **tf.reduce_mean** (valor médio), **tf.reduce_all** (E lógico), **tf.reduce_any** (OU lógico) e **tf.reduce_logsumexp** (log(sum(exp))).

Os métodos para usar essas operações são semelhantes. A seguir descreve como usar **tf.reduce_sum**. Calcula a soma dos elementos em cada dimensão de um tensor **(tf.reduce_sum(input_tensor, axis=None, keepdims=False,name=None))**:
*   **input_tensor**: tensor de entrada.
*   **axis**: especifica o eixo a ser calculado. Se este parâmetro não for especificado, o valor médio de todos os elementos é calculado.
*   **keepdims**: especifica se as dimensões devem ser mantidas. Se este parâmetro estiver definido como **True**, o o resultado de saída mantém a forma do tensor de entrada. Se este parâmetro estiver definido como **False**, as dimensões do resultado de saída diminuem.
*   **name**: nome da operação.

```python
reduce_sample_1 = tf.constant([1,2,3,4,5,6],shape=[2,3])
print("calcular a soma de todos os elementos no tensor (axis=None):",tf.reduce_sum(reduce_sample_1,axis=None).numpy())
print("calcular a soma dos elementos em cada coluna por coluna (axis=0):",tf.reduce_sum(reduce_sample_1,axis=0).numpy())
print("calcular a soma dos elementos em cada coluna por linha (axis=1):",tf.reduce_sum(reduce_sample_1,axis=1).numpy())
```

####**1.2.1.7 Concatenação e divisão de tensores**

####**1.2.1.7.1 Concatenação de Tensores**

No TensorFlow, as operações de concatenação de tensores incluem:
*   **tf.contact()**: concatena vetores com base na dimensão especificada, mantendo outras dimensões inalteradas.
*   **tf.stack()**: altera um grupo de tensores dimensionais R para tensores dimensionais R+1, com as dimensões alteradas após a concatenação.

**tf.concat(values, axis, name='concat')**:
*   **values**: valores para concatenação de tensores
*   **axis**: dimensão a concatenar
*   **name**: nome da operação

```python
concat_sample_1 = tf.random.normal([4,100,100,3])
concat_sample_2 = tf.random.normal([40,100,100,3])
print("tamanhos dos dados originais: ",concat_sample_1.shape, concat_sample_2.shape)
concated_sample_1 = tf.concat([concat_sample_1,concat_sample_2],axis=0)
print("tamanho dos dados concatenados:",concated_sample_1.shape)
```

Uma dimensão é adicionada à matriz original. Da mesma forma, o eixo dos parâmetros determina a posição da dimensão.
**tf.stack(values, axis=0, name='stack')**:
*   **values**: tensores de entrada, um grupo de tensores com a mesma forma e tipo de dados.
*   **axis**: dimensão a concatenar.
*   **name**: nome da operação.

```python
stack_sample_1 = tf.random.normal([100,100,3])
stack_sample_2 = tf.random.normal([100,100,3])
print("tamanhos dos dados originais: ",stack_sample_1.shape, stack_sample_2.shape)
#Dimensões aumentam após a concatenação. Se o eixo for definido como 0, uma dimensão será adicionada antes da primeira dimensão.
stacked_sample_1 = tf.stack([stack_sample_1, stack_sample_2],axis=0)
print("tamanho dos dados concatenados:",stacked_sample_1.shape)
```

Uma dimensão é adicionada à matriz original. Da mesma forma, o eixo dos parâmetros determina a posição da dimensão.

**tf.stack(values, axis=0, name='stack')**:

#####**1.2.1.7.2 Divisão de Tensores**

No TensorFlow, as operações de divisão de tensor incluem:
*   **tf.unstack()**: divide um tensor por uma dimensão específica.
*   **tf.split()**: divide um tensor em um número especificado de subtensores com base em uma dimensão específica.

**tf.split()** é mais flexível que **tf.unstack()**.
**tf.unstack(value,num=None,axis=0,name='unstack')**:
*   **value**: tensor de entrada
*   **num**: indica que uma lista contendo **num** elementos é gerada. O valor de num deveser igual ao número de elementos na dimensão especificada. Este parâmetro geralmente pode ser ignorado.
*   **axis**: indica a dimensão com base na qual o tensor é dividido.
*   **name**: nome da operação

```python
#Divida os dados com base na primeira dimensão e produza os dados divididos em uma lista.
tf.unstack(stacked_sample_1,axis=0)
```

**tf.split(value, num_or_size_splits, axis=0)**:
*   **value**: tensor de entrada
*   **num_or_size_splits**: número de subtensores
*   **axis**: indica a dimensão com base na qual o tensor é dividido.

**tf.split()** divide um tensor de uma das seguintes maneiras:
1.   Se o valor de **num_or_size_splits** for um número inteiro, o tensor será dividido igualmente em sub-tensores na dimensão especificada (eixo = D).
2.   Se o valor de **num_or_size_split**s for um vetor, o tensor é dividido em subtensores com base no valor do elemento do vetor na dimensão especificada (eixo = D)

```python
import numpy as np
split_sample_1 = tf.random.normal([10,100,100,3])
print("tamanho dos dados originais:",split_sample_1.shape)
splited_sample_1 = tf.split(split_sample_1, num_or_size_splits=5,axis=0)
print("tamanho dos dados divididos quando m_or_size_splits é definido como 10: ",np.shape(splited_sample_1))
splited_sample_2 = tf.split(split_sample_1, num_or_size_splits=[3,5,2],axis=0)
print("Quando num_or_size_splits=[3,5,2], o tamanho dos dados divididos é:", np.shape(splited_sample_2[0]),
np.shape(splited_sample_2[1]), np.shape(splited_sample_2[2]))
```

#####**1.2.1.8 Ordenação de tensores**

No TensorFlow, as operações de ordenação de tensores incluem:
*   **tf.sort()**: ordena os tensores em ordem crescente ou decrescente e retorna os tensores ordenados.
*   **tf.argsort()**: ordena os tensores em ordem crescente ou decrescente e retorna os índices dos tensores.
*   **tf.nn.top_k()**: retorna os primeiros k valores máximos.
**tf.sort/argsort(input, direction, axis)**:
*   **input**: tensor de entrada
*   **direction**: ordem de classificação, que pode ser definida como **DESCENDING** (ordem decrescente) ou **ASCENDING** (ordem crescente). O valor padrão é **ASCENDING**.
*   **axis**: classificação pela dimensão especificada pelo eixo. O valor padrão do eixo é –1, indicando a última dimensão.

```python
sort_sample_1 = tf.random.shuffle(tf.range(10))
print("tensor de entrada:",sort_sample_1.numpy())
sorted_sample_1 = tf.sort(sort_sample_1, direction="ASCENDING")
print("tensor classificado em ordem crescente:",sorted_sample_1.numpy())
sorted_sample_2 = tf.argsort(sort_sample_1,direction="ASCENDING")
print("índices dos elementos dos tensores em ordem crescente:",sorted_sample_2.numpy())
```

###**1.2.2 Modo Eager Execution do TensorFlow 2**

**Modo Eager Execution**:
O modo Eager Execution do TensorFlow é um tipo de programação imperativa, que é o mesmo que Python nativo. Quando você executa uma operação específica, o sistema retorna imediatamente um resultado.

**Modo Graph Execution**:
O TensorFlow 1.0 adota o modo Graph Execution para primeiro construir um gráfico computacional, habilitar uma Sessão e, em seguida, alimentar os dados reais para obter um resultado. No modo Eager Execution, a depuração de código é mais fácil, mas a eficiência de execução do código é menor. A seguir implementamos multiplicação simples usando TensorFlow para comparar as diferenças entre o modo de execução Eager  e modo Graph.

```python
x = tf.ones((2, 2), dtype=tf.dtypes.float32)
y = tf.constant([[1, 2], [3, 4]], dtype=tf.dtypes.float32)
z = tf.matmul(x, y)
print(z)
```

```python
#Use a sintaxe do TensorFlow 1.x no TensorFlow 2. Você pode instalar o pacote de compatibilidade v1 no TensorFlow 2.0 para herdar o código do TensorFlow 1.x e desativar o modo de execução antecipado.
import tensorflow.compat.v1 as tf
tf.disable_eager_execution()
#Crie um grafo e defina-o como um grafo computacional.
a = tf.ones((2, 2), dtype=tf.dtypes.float32)
b = tf.constant([[1, 2], [3, 4]], dtype=tf.dtypes.float32)
c = tf.matmul(a, b)
#Ative a função de desenho e execute a operação de multiplicação para obter os dados.
with tf.Session() as sess:
  print(sess.run(c))
```

Reinicie o kernel para restaurar o TensorFlow 2.0 e ativar o modo de execução eager. Outra vantagem do modo Eager Execution está na disponibilidade de funções nativas do Python, por
exemplo, instruções condicional:

```python
import tensorflow as tf
thre_1 = tf.random.uniform([], 0, 1)
x = tf.reshape(tf.range(0, 4), [2, 2])
print(thre_1)
if thre_1.numpy() > 0.5:
  y = tf.matmul(x, x)
else:
  y = tf.add(x, x)
```

###**1.2.3 TensorFlow2 Autograph**

Ao ser usado para comentar uma função, o decorador **tf.function** pode ser chamado como qualquer outra função. **tf.function** será compilado em um grafo, para que possa rodar mais eficientemente em uma GPU ou TPU. Nesse caso, a função se torna uma operação no TensorFlow. A função pode ser chamada diretamente para gerar um valor de retorno. Contudo, a função é
executada no modo graph e os valores das variáveis intermediárias não podem ser visualizados diretamente.

```python
@tf.function
def simple_nn_layer(w,x,b):
  print(b)
  return tf.nn.relu(tf.matmul(w, x)+b)
w = tf.random.uniform((3, 3))
x = tf.random.uniform((3, 3))
b = tf.constant(0.5, dtype='float32')
simple_nn_layer(w,x,b)
```

De acordo com a saída do comando, o valor de **b** na função não pode ser visualizado diretamente,
mas o valor de retorno pode ser visualizado usando **.numpy()**. A seguir, comparamos o
desempenho do modo graph e modo de execução eager, executando a mesmo operação (cálculo de uma camada CNN)

```python
#Use o módulo timeit para medir o tempo de execução de um pequeno segmento de código.
import timeit
#Crie uma camada convolucional.
CNN_cell = tf.keras.layers.Conv2D(filters=100,kernel_size=2,strides=(1,1))
#Use @tf.function para converter a operação em modo graph.
@tf.function
def CNN_fn(image):
  return CNN_cell(image)
image = tf.zeros([100, 200, 200, 3])
#Compare the execution time of the two modes.
CNN_cell(image)
CNN_fn(image)
#Chame timeit.timeit para medir o tempo necessário para executar o código 10 vezes.
print("tempo necessário para realizar o cálculo de uma camada de rede neural convolucional (CNN) no modo de execução eager:", timeit.timeit(lambda: CNN_cell(image), number=10))
print("tempo necessário para realizar o cálculo de uma camada CNN em modo graph:", timeit.timeit(lambda: CNN_fn(image), number=10))
```

Por exemplo, para uma Imagem RGB 128 x 128, os dados são organizados como (3, 128, 128) se o valor for **channels_first** e (128, 128, 3) se o valor for **channels_last**. O valor padrão deste parâmetro é o valor definido em **~/.keras/keras.json**. Se o valor não for definido, o o valor é **channels_last**.

####**2.2.2.1 Compilando**

Após a construção de um modelo, você pode chamar **compile** para configurar o processo de aprendizagem do modelo:

####**2.2.2.2 Treinamento**

**fit(x=None, y=None, batch_size=None, epochs=1, verbose=1, callbacks=None, validation_split=0.0, validation_data=None, shuffle=True, class_weight=None, sample_weight=None, initial_epoch=0, steps_per_epoch=None, validation_steps=None)**:
*   **x**: Insira dados de treinamento
*   **y**: Target (dados rotulados)
*   **batch_size**: Número de amostras para cada atualização de gradiente. O valor padrão é **32**.
*   **epochs**: número de rodadas de iteração do modelo de treinamento.
*   **verbose**: modo de exibição de logs, definido como **0, 1** ou **2**. 0 = Não exibido, 1 = Barra de progresso, 2 =
Uma linha é exibida para cada rodada.
*   **callbacks**: função de retorno de chamada usada durante o treinamento.
*   **validation_split**: fração dos dados de treinamento a serem usados como dados de validação.
*   **validation_data**: conjunto de validação. Este parâmetro substituirá **validação_split**.
*   **shuffle**: indica se os dados devem ser embaralhados antes de cada rodada de iteração. Esse  parâmetro é inválido quando **steps_per_epoch** não é **None**.
*   **initial_epoch**: época para iniciar o treinamento (útil para retomar um treinamento de pesos anterior).
*   **steps_per_epoch**: defina como o tamanho do conjunto de dados ou **batch_size**.
*   **validation_steps**: Este parâmetro é válido somente quando **steps_per_epoch** é especificado. Número total de etapas (lotes de amostras) a validar antes de parar.

```python
import numpy as np
train_x = np.random.random((1000, 36))
train_y = np.random.random((1000, 10))
val_x = np.random.random((200, 36))
val_y = np.random.random((200, 10))
model.fit(train_x, train_y, epochs=10, batch_size=100,validation_data=(val_x, val_y))
```

Você pode usar **tf.data** para criar pipelines de entrada de treinamento para grandes conjuntos de dados.

```python
dataset = tf.data.Dataset.from_tensor_slices((train_x, train_y))
dataset = dataset.batch(32)
dataset = dataset.repeat()
val_dataset = tf.data.Dataset.from_tensor_slices((val_x, val_y))
val_dataset = val_dataset.batch(32)
val_dataset = val_dataset.repeat()
model.fit(dataset, epochs=10, steps_per_epoch=30,validation_data=val_dataset, validation_steps=3)
```

Uma função de retorno de chamada (**callback**) é um objeto passado ao modelo para personalizar e estender o comportamento do modelo durante o treinamento. Podemos escrever nossos próprios **callbacks** personalizados ou usar funções integradas em **tf.keras.callbacks**. As funções **callbacks** comuns são as seguintes:
*   **tf.keras.callbacks.ModelCheckpoint**: salva modelos periodicamente.
*   **tf.keras.callbacks.LearningRateScheduler**: altera dinamicamente a taxa de aprendizagem.
*   **tf.keras.callbacks.EarlyStopping**: interrompe o treinamento antecipadamente.
*   **tf.keras.callbacks.TensorBoard**: usa TensorBoard.

```python
import os
Epochs = 10
logdir=os.path.join("logs")
if not os.path.exists(logdir):
  os.mkdir(logdir)
# Defina uma função para definir dinamicamente a taxa de aprendizagem.
def lr_Scheduler(epoch):
  if epoch > 0.9 * Epochs:
    lr = 0.0001
  elif epoch > 0.5 * Epochs:
    lr = 0.001
  elif epoch > 0.25 * Epochs:
    lr = 0.01
  else:
    lr = 0.1
  print(lr)
  return lr
callbacks = [
  #Parada antecipada:
  tf.keras.callbacks.EarlyStopping(
    #Métrica para determinar se o desempenho do modelo não apresenta melhorias adicionais.
    monitor='val_loss',
    #Limite para determinar se o desempenho do modelo não apresenta melhorias adicionais.
    min_delta=1e-2,
    #Número de épocas em que o desempenho do modelo não apresenta melhorias adicionais.
    patience=2),
    #Salve modelos periodicamente:
    tf.keras.callbacks.ModelCheckpoint(
      #Caminho do modelo
      filepath='testmodel_{epoch}.h5',
      #Se salvar o modelo ideal
      save_best_only=True,
      #Métrica monitorada
      monitor='val_loss'),
    #Altere dinamicamente a taxa de aprendizagem.
    tf.keras.callbacks.LearningRateScheduler(lr_Scheduler),
    #Use o TensorBoard.
    tf.keras.callbacks.TensorBoard(log_dir=logdir)
]
model.fit(train_x, train_y, batch_size=16, epochs=Epochs,callbacks=callbacks, validation_data=(val_x,val_y))
```

O parâmetro **num_classes=10** indica que o conjunto de dados consiste em 10 classes.


Passo 2 - Divida amostras em lotes e classifique-as em pequenos lotes.

Mapeie variáveis independentes para variáveis dependentes do conjunto de dados e, em seguida, classifique as amostras em pequenos lotes para melhorar o desempenho do  gradiente descendente. Neste exemplo, 50 amostras são classificadas em um grupo.

```python
train_input=tf.data.Dataset.from_tensor_slices((train_X,train_label)).batch(50)
test_input=tf.data.Dataset.from_tensor_slices((test_X,test_label)).batch(50)
```

###**3.2.3 Construindo uma Rede Neural**

Passo 1 - Definindo uma camada de entrada.

Especifique o número de neurônios na camada de entrada. No entanto, como os dados são carregados no formato de matriz no início, a camada de entrada é representada no formato de matriz.

```python
input_data=tf.keras.Input([28,28])
```

Passo 2 - Achate (**flatten**) os dados.
Achate os dados para representá-los no formato vetorial.

```python
dense=tf.keras.layers.Flatten()(input_data)
```

Passo 3 - Definindo a camada intermediária (camada oculta).

```python
dense_1=tf.keras.layers.Dense(100,activation='relu')(dense)
dense_2=tf.keras.layers.Dense(100,activation='relu')(dense_1)
dense_3=tf.keras.layers.Dense(100,activation='relu')(dense_2)
dense_4=tf.keras.layers.Dense(100,activation='relu')(dense_3)
```

Neste exemplo, 100 neurônios são definidos para cada camada, e a função de ativação **ReLU** é usada.

Passo 4 - Definindo a camada de saída.

Como são especificadas dez classes, são necessários dez neurônios na camada de saída, e a função ativação softmax
é usada.

```python
output_data=tf.keras.layers.Dense(10,activation='softmax')(dense_4)
```

Passo 5 - Definindo a sequência do modelo.

O processo geral de construção do modelo está concluído até agora. Depois, definimos a entrada e sequências de saída do modelo.

```python
model=tf.keras.Model(inputs=input_data,outputs=output_data)
```

Passo 6 - Definindo a função de perda, o otimizador e as informações de saída de treinamento.

```python
model.compile(optimizer=tf.keras.optimizers.legacy.Adam(learning_rate=0.001),loss=tf.losses.categorical_crossentropy, metrics=['accuracy'])
```

Foi especificado o otimizador **Adam**, definida a taxa de aprendizado como **0,001** e a função de perda como entropia cruzada.
Posteriormente, produza a taxa de acuracia do modelo atual após cada rodada de treinamento (**metrics=['accuracy']**).

Passo 7 - Visualizando o modelo.

```python
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True)
```

Visualize o modelo declarado verificando a quantidade de parâmetros a serem treinados em cada camada,
o número de parâmetros a serem treinados em toda a rede e o número de neurônios
em cada camada.

Executa **10**rodadas de treinamento.

Passo 9 Avalie o modelo usando o conjunto de testes.

Execute a avaliação de predição usando o conjunto de testes para determinar se o modelo é apropriado.


```python
model.evaluate(test_input)
```

###**3.2.4 Criando uma CNN**
Durante a construção da CNN, o pré-processamento de dados leva um tempo relativamente longo porque o imagens pretas e brancas são usadas (há apenas um canal de cores). No entanto, o conjunto de dados não é definido. Portanto, você precisa adicionar uma dimensão de forma proativa.


```python
import tensorflow as tf
fashion_mnist=tf.keras.datasets.fashion_mnist
(train_X,train_y),(test_X,test_y)=fashion_mnist.load_data()
train_data=tf.expand_dims(train_X,-1)
test_data=tf.expand_dims(test_X,-1)
```

Em seguida, execute a costura e o processamento em pequenos lotes.

```python
train_input=tf.data.Dataset.from_tensor_slices((train_data,train_label)).batch(50)
test_input=tf.data.Dataset.from_tensor_slices((test_data,test_label)).batch(50)
```

Passo 1 - Criando a primeira camada, que possui uma camada convolucional 2D com tamanho de kernel 3x3 e operação de max pooling

```python
from tensorflow.python import keras
from tensorflow.python.keras.models import Sequential
from keras.layers import Dense, Conv2D, Activation, MaxPool2D, Flatten, Dropout, BatchNormalization
from keras.optimizers import RMSprop,Adam
model = tf.keras.Sequential()

# Primeira camada
model.add(Conv2D(32, (3,3), padding='same', input_shape=(28,28, 1)))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
```

Passo 2 - Segunda camada, que possui uma camada convolucional 2D com tamanho de kernel como ativação 3x3 e ReLU e operação de pooling máximo

```python
model.add(Conv2D(64, (3,3), padding='same', activation=tf.nn.relu))
model.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2)))
```

Passo 3 - Camada totalmente conectada com função de ativação ReLU

```python
model.add(Flatten())
model.add(Dense(128, activation=tf.nn.relu))
```

Passo 4 - Camada de saída com função de ativação softmax

```python
# Output layer with softmax activation function
model.add(Dense(10, activation=tf.nn.softmax))
model.summary()
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True, show_layer_names=True)
```

Passo 5 - Compilando e definindo a função de perda, o otimizador e as informações de saída de treinamento.

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.categorical_crossentropy,
              metrics=['accuracy'])
```

Passo 6 - Treine e teste o modelo.

```python
model.fit(train_input,epochs=15)
```

```python
model.evaluate(test_input)
```
