##**2.2 Comando if**

Você pode usar a instrução if em Python para determinar o nível de uma pontuação inserida por um usuário.

```python
# Determine a pontuação inserida.
# input(): Recebe dados de entrada.
score = input("Por favor, insira sua pontuação.") # A função de entrada recebe uma entrada, que é uma string.
# try: except Exception: ... é uma instrução Python usada para capturar exceções. Se ocorrer um erro no bloco da instrução em try,
# a instrução em except é executada.
try:
  score = float(score) # Converte a pontuação em um número.
  if 100>=score>=90: # Verifica se o valor inserido é maior que a pontuação de um nível.
    print("Excelente") # Gera o nível quando as condições são conheceu.
  elif 90 > score >= 80:
    print("Bom")
  elif 80>score>60:
    print("Médio")
  else:
    print("Ruim")
except Exception:
  print("Insira uma pontuação correta.")
```

##**2.2 Comandos de repetição (loop)**

Usando o loop for para gerar uma tabuada.

Passo 1 - loop while

Quando a condição é atendida, o bloco de instruções é executado ciclicamente. Para finalizar o loop, use **break** ou **continue**.

```python
# loop while
i = 0 # Crie a variável i.
while i<9:
  # Define uma condição para o loop.
  i+=1 # O valor de i aumenta em 1 em cada loop.
  if i == 3: # Verifique se as condições foram atendidas.
    print("Sair deste loop.")
    continue # Execute continue para sair do loop atual.
  if i == 6:
    print("Sair do grande loop atual.")
    break # Sai do grande loop atual.
  print(i)

print("loop terminou.")
```

##**3.1 Personalização de funções**

Passo 1 - Defina uma função.

Defina uma função para retornar uma sequência. Cada número na sequência é a soma dos dois números anteriores (sequência de Fibonacci).

```python
def fibs(num): # Parâmetro de posição.
  resultado = [0,1] # Crie uma lista para armazenar o valor da sequência.
  for i in range(2,num): # Ciclo num-2 vezes.
    a = resultado[i-1] + resultado[i-2]
    resultado.append(a) # Anexa o valor à lista.
  return resultado # Retorna a lista.
fibs(5)
```

Crie a classe Cachorro. Cada instância criada com base na classe Cachorro armazena o nome e a idade. Daremos a cada Cachorro as habilidades **sentar()** e **rolar()**.

```python
class Cachorro():
  """Uma simples tentativa de simular um cachorro"""
  def __init__ (self,nome,idade):
    """Inicialize os atributos nome e idade."""
    self.name = nome
    self.idade = idade
  def sentar(self):
    """Simule um cachorro sentado quando solicitado."""
    print(self.name+" agora está sentado")
  def rolar(self):
    """Simule um cachorro rolando quando solicitado."""
    print(self.name+" rolou!")
# Instancia uma classe.
cachorro = Cachorro("Husky",2)
cachorro.sentar()
cachorro.rolar()
```

##**6.1 Expressão Regular**

Passo 1 - função **re.match**:

A função re.match tenta corresponder a um padrão da posição inicial da string. Se a correspondencia não foi bem-sucedida, match() retorna none.

Sintaxe da função:
**re.match(pattern, string, flags=0)**

```python
import re
print(re.match('www', 'www.huawei.com').span()) # Corresponde à posição inicial.
print(re.match('com', 'www.huawei.com')) # Não corresponde à posição inicial.
```

Passo 2 - Método **re.search**:

O método re.search verifica toda a string e retorna a primeira correspondência bem-sucedida.
Sintaxe da função: **re.search(pattern, string, flags=0)**

```python
import re
linha = "Gatos são mais espertos que cachorros"
searchObj = re.search( r'(.*) são (.*?) .*', linha, re.M|re.I)
if searchObj:
  print("searchObj.group(): ", searchObj.group())
  print("searchObj.group(1) : ", searchObj.group(1))
  print("searchObj.group(2) : ", searchObj.group(2))
else:
  print("Nada encontrado!!")
```

Passo 3 - Recuperação e substituição:

O módulo re do Python fornece a função re.sub para substituir os itens correspondentes na string.

Sintaxe: **re.sub(pattern, repl, string, count=0, flags=0)**

```python
import re
telefone = "95-90101-00077 # Este é um número de telefone."
# Exclua o comentário Python na string.
num = re.sub(r'#.*$', "", telefone)
print("O número do telefone é: ", num)
# Exclua os hífens do número de telefone.
num = re.sub(r'\D', "", telefone)
print("O número do telefone é: ", num)
```

Passo 4 - A função **re.compile**:

A função compile é usada para compilar expressões regulares e gerar um regular objeto de expressão (padrão) para as funções match() e search().
O formato da sintaxe é o seguinte: **re.compile(pattern[, flags])**

```python
import re
pattern = re.compile(r'\d+') # Pelo menos um dígito é correspondido.
n = pattern.match('one12twotrês34four') # O cabeçalho não foi encontrado.
print(n)
m = pattern.search('one12twotrês34four') # Corresponde a partir da posição de 'e'. Nenhuma correspondência foi encontrada.
print(m)
print(m.group())
```

Passo 2 - Criar um gerador por meio de uma palavra-chave **yield**.

Use a palavra-chave **yield** para criar um gerador para gerar a sequência de Fibonacci

```python
#revisar yeld
def fib(n):
  current = 0
  num1, num2 = 0, 1
  while current < n:
    num = num1
    num1, num2 = num2, num1+num2
    current += 1
    yield num
  return 'done'
g=fib(5)
while True:
  try:
    x = next(g)
    print("value:%d"%x)
  except StopIteration as e:
    print("Generator return value:%s"%e.value)
    break
```

Passo 3 - Usando o método **send**.

Além do método **next**, o método **send **pode ser usado para ativar o gerador.
Comparado com o método **next**, o método **send** também pode transferir dados para o ponto de interrupção ao acordar o gerador.

```python
def gen():
  i = 0
  while i<5:
    temp = yield i
    print(temp)
    i+=1
f = gen()
next(f)
```

##**7.3 Decoradores**

Passo 1 - Construindo um decorador.

Crie um decorador simples que calcule o tempo de execução de uma função.

```python
import time
def func(f):
  def inner(*args, **kwargs):
    start_time = time.time()
    f(*args, **kwargs)
    end_time = time.time()
    print('Consumed time:%s second' % (end_time - start_time))
  return inner
```

Passo 2 - Usando a função de decoração.

```python
@func
def test():
  time.sleep(2)
test()

```
