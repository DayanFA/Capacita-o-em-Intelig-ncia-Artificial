##**2.2 Comando if**

Voc pode usar a instruo if em Python para determinar o nvel de uma pontuao inserida por um usurio.

```python
# Determine a pontuao inserida.
# input(): Recebe dados de entrada.
score = input("Por favor, insira sua pontuao.") # A funo de entrada recebe uma entrada, que  uma string.
# try: except Exception: ...  uma instruo Python usada para capturar excees. Se ocorrer um erro no bloco da instruo em try,
# a instruo em except  executada.
try:
  score = float(score) # Converte a pontuao em um nmero.
  if 100>=score>=90: # Verifica se o valor inserido  maior que a pontuao de um nvel.
    print("Excelente") # Gera o nvel quando as condies so conheceu.
  elif 90 > score >= 80:
    print("Bom")
  elif 80>score>60:
    print("Mdio")
  else:
    print("Ruim")
except Exception:
  print("Insira uma pontuao correta.")
```

Quando a condio  atendida, o bloco de instrues  executado ciclicamente. Para finalizar o loop, use **break** ou **continue**.

```python
# loop while
i = 0 # Crie a varivel i.
while i<9:
  # Define uma condio para o loop.
  i+=1 # O valor de i aumenta em 1 em cada loop.
  if i == 3: # Verifique se as condies foram atendidas.
    print("Sair deste loop.")
    continue # Execute continue para sair do loop atual.
  if i == 6:
    print("Sair do grande loop atual.")
    break # Sai do grande loop atual.
  print(i)

print("loop terminou.")
```

##**3.1 Personalizao de funes**

Passo 1 - Defina uma funo.

Defina uma funo para retornar uma sequncia. Cada nmero na sequncia  a soma dos dois nmeros anteriores (sequncia de Fibonacci).

```python
def fibs(num): # Parmetro de posio.
  resultado = [0,1] # Crie uma lista para armazenar o valor da sequncia.
  for i in range(2,num): # Ciclo num-2 vezes.
    a = resultado[i-1] + resultado[i-2]
    resultado.append(a) # Anexa o valor  lista.
  return resultado # Retorna a lista.
fibs(5)
```

Passo 2 - Mtodo **re.search**:

O mtodo re.search verifica toda a string e retorna a primeira correspondncia bem-sucedida.
Sintaxe da funo: **re.search(pattern, string, flags=0)**

```python
import re
linha = "Gatos so mais espertos que cachorros"
searchObj = re.search( r'(.*) so (.*?) .*', linha, re.M|re.I)
if searchObj:
  print("searchObj.group(): ", searchObj.group())
  print("searchObj.group(1) : ", searchObj.group(1))
  print("searchObj.group(2) : ", searchObj.group(2))
else:
  print("Nada encontrado!!")
```

Passo 3 - Recuperao e substituio:

O mdulo re do Python fornece a funo re.sub para substituir os itens correspondentes na string.

Sintaxe: **re.sub(pattern, repl, string, count=0, flags=0)**

```python
import re
telefone = "95-90101-00077 # Este  um nmero de telefone."
# Exclua o comentrio Python na string.
num = re.sub(r'#.*$', "", telefone)
print("O nmero do telefone : ", num)
# Exclua os hfens do nmero de telefone.
num = re.sub(r'\D', "", telefone)
print("O nmero do telefone : ", num)
```

##**7.2 Geradores**

Passo 1 - Criar um gerador por meio de derivao de lista.

```python
G = (x*2 for x in range(5))
print(type(G))
print(*G)
```

Passo 2 - Criar um gerador por meio de uma palavra-chave **yield**.

Use a palavra-chave **yield** para criar um gerador para gerar a sequncia de Fibonacci

```python
#revisar yeld
def fib(n):
  current = 0
  num1, num2 = 0, 1
  while current < n:
    num = num1
    num1, num2 = num2, num1+num2
    current += 1
    yield num
  return 'done'
g=fib(5)
while True:
  try:
    x = next(g)
    print("value:%d"%x)
  except StopIteration as e:
    print("Generator return value:%s"%e.value)
    break
```

Passo 3 - Usando o mtodo **send**.

Alm do mtodo **next**, o mtodo **send **pode ser usado para ativar o gerador.
Comparado com o mtodo **next**, o mtodo **send** tambm pode transferir dados para o ponto de interrupo ao acordar o gerador.

```python
def gen():
  i = 0
  while i<5:
    temp = yield i
    print(temp)
    i+=1
f = gen()
next(f)
```

##**7.3 Decoradores**

Passo 1 - Construindo um decorador.

Crie um decorador simples que calcule o tempo de execuo de uma funo.

```python
import time
def func(f):
  def inner(*args, **kwargs):
    start_time = time.time()
    f(*args, **kwargs)
    end_time = time.time()
    print('Consumed time:%s second' % (end_time - start_time))
  return inner
```

Passo 2 - Usando a funo de decorao.

```python
@func
def test():
  time.sleep(2)
test()

```
